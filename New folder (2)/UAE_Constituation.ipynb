Initially, I extracted all of the text data.Basically, everything connected to the article.The pdf file was then converted into a csv file.


```python
import PyPDF2
import csv

# Open the PDF file
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/UAE_Labour_Law.pdf', 'rb') as file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(file)

    # Initialize an empty list to store the extracted articles
    articles = []


    # Initialize a variable to keep track of the current article content
    #current_article = ""


    # Iterate over each page in the PDF
    for page in pdf_reader.pages:


        # Extract the text from the page
        text = page.extract_text()

        # Split the text into lines
        lines = text.split('\n')

        # Iterate over each line
        for line in lines:
            # Check if the line starts with 'Article'
            if line.startswith('Article'):
                # If we have accumulated content for a previous article, add it to the list
                if current_article:
                    articles.append(current_article)
                # Start a new article
                current_article = line
            else:
                # If it's not a line starting with 'Article', add it to the current article content
                current_article += "\n" + line

    # Add the last article (if any) to the list
    if current_article:
        articles.append(current_article)

# Prepare the data for CSV conversion
data = [[article] for article in articles]

# Convert the data to CSV
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

I preprocessed the dataset using stopwords, punctuation, lowercase letters, and tokenization from the new csv file.

```python
import csv
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Function to perform text preprocessing
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove punctuation and numbers
    tokens = [word for word in tokens if word.isalpha()]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    return ' '.join(tokens)

# Read the CSV file and preprocess the text
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
    preprocessed_data = [[preprocess_text(article[0])] for article in data]

# Write the preprocessed data back to a new CSV file
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles_preprocessed.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(preprocessed_data)
```

After removing the first row, I found that because it only has one column, it will not be precise enough to generate an authenticate answer.That is why I chose the BART model.That is, generate summarized text from each row and save it with that row.  


```python
from transformers import BartTokenizer, BartForConditionalGeneration
import csv

# Load BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Function to generate summary
def generate_summary(text):
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs.input_ids, num_beams=4, length_penalty=2.0, max_length=150, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Read the CSV file and generate summaries
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles_summarized.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
    summaries_data = []

    for idx, article in enumerate(data):
        text = article[0]
        summary = generate_summary(text)
        summaries_data.append([summary, text])

# Write the summaries back to a new CSV file
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles_summarized_new.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(summaries_data)
```

Following that, I added headers to two columns: the first is a title generated by the BART model, and the second is text.

```python
# Read the CSV file and add headers
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles_summarized_new.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
    data_with_headers = [["title", "text"]] + data  # Add headers

# Write the data back to the CSV file with headers
with open('/content/drive/MyDrive/Machine_Learning/Inflexion/articles_summarized_new_one.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data_with_headers)
```
